```python
df = pd.read_csv('./tests/all.csv')
```

```python
df
```

```python
etf = df[df['品种']=='好买'].iloc[0]['余额'] - df[df['品种']=='股票型主动基金'].iloc[0]['余额'] + \
      df[df['品种']=='且慢'].iloc[0]['余额'] - df[df['品种']=='我要稳稳的幸福'].iloc[0]['余额']
#且慢只能统计累积总收益,年初累计为20425.
etf_add = df[df['品种']=='且慢'].iloc[0]['当前收益'] -20425- df[df['品种']=='我要稳稳的幸福'].iloc[0]['当前收益']+df[df['品种']=='好买'].iloc[0]['当前收益'] - df[df['品种']=='股票型主动基金'].iloc[0]['当前收益']

df.loc[df.index.max() + 1] = ['etf',etf,0,etf_add]
```

```python
df
```

```python
total = df.iloc[0]['余额'] + df.iloc[1]['余额'] + df.iloc[2]['余额'] + df.iloc[3]['余额'] + df.iloc[6]['余额']

df.loc[df.index.max() + 1] = ['总额',total,0,0]
df['收益率'] = df['当前收益']/(df['余额']-df['当前收益'])
df['收益率'] = df['收益率'].map('{:,.3f}'.format)
df['市值占比']= df['余额']/total
```

```python
df = df.drop(index=[4,5],axis=0)
df.sort_values(by=['市值占比'],ascending=False)
```
